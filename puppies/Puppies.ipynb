```python

import numpy as np
import pandas as pd
import matplotlib as plt
import re
from datetime import date
def days_on_site(d1):
    d0=pd.to_datetime('2016/4/02') 
    delta = d0 - d1
    return delta.days

df1 = pd.read_csv('pupfile.42.xls','\t') 
df1.columns = ['Posted Date', 'Title','type','Breed','Locale','Gender', 'Age_Cat', 'Description']
df1.head()

```

```python
df1.count()
# how to name the columns in my pandas dataframe????

# how to change all values in a column to something else
df1['Posted Date'] = pd.to_datetime(df1['Posted Date'])


```

```python
df1.head()
```

```python
df=pd.DataFrame(df1.Locale.str.split(',',1).tolist(),
                                   columns = ['city','state'])
df2=pd.concat([df1,df],axis=1)
df2.head(10)
```

```python
age = []
gender = []
rescue = []
days = []
for i in range(0,df2.Description.count()):
    a= df2.Description[i]
    if a.find("Posted Breed:",0,13) == 0:
        index=a.find(".",12)
        df2.Description.loc[i] = a[index+1:].strip()
    # extract age (best guess)
    r=re.search(r'([0-9]+)(?= (yrs|years|year) old)',df2.Description.loc[i],re.IGNORECASE)
    if r and int(r.group(0))<20:
        age.append( int(r.group(0)))
    else:
        age.append(None)
        
    # set gender dummy variable
    gender.append(1 if df2.Gender.loc[i]=="Male" else 0)
    # set rescue dummy variable
    rescue.append(1 if df2.type.loc[i]=="rescue" else 0)
    days.append(days_on_site(df2['Posted Date'].loc[i]))
    
df2['age'] = age
df2['is_male'] = gender
df2['is_rescue'] = rescue
df2['days_on_site'] = days

df2['days_on_site'].describe()

```

```python
############
import re
from textblob import TextBlob
sentiment=[]
subjectivity=[]
for a in df2.Description:
    if not pd.isnull(a):
        b=a.decode('utf8')
        m=TextBlob(b)
        sentiment.append(m.sentiment.polarity)
        subjectivity.append(m.sentiment.subjectivity)
    else:
        sentiment.append(None)
        subjectivity.append(None)

batch = '402'

df2['sentiment']=sentiment
df2['subjectivity']=subjectivity
df2.columns.values
```

```python
df2['batch'] = batch
df2.to_csv(path_or_buf='pupfile_ready'+batch+'.xls', sep='\t', na_rep='')
df4=pd.read_csv('pupfile_ready'+batch+'.xls','\t') 
```
